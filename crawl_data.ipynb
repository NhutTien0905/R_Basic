### Crawl data from web

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

# Initialize WebDriver
driver = webdriver.Chrome()

# Navigate to initial page 
# driver.get("https://s.cafef.vn/lich-su-giao-dich-ree-1.chn#data") #HSG  02/06/2000
# driver.get("https://s.cafef.vn/lich-su-giao-dich-bbc-1.chn#data") #BID 19/12/2001
driver.get("https://s.cafef.vn/lich-su-giao-dich-piv-1.chn#data") #HPG 15/12/2008

date_time_string = "19/08/2010 - 08/09/2023"
date_input = driver.find_element(By.ID, "date-inp-disclosure")
date_input.send_keys(date_time_string)

while True:
    # Click the Download button
    download_button = driver.find_element(By.ID, "tabletoExcel")
    download_button.click()

    # Crawl data using BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Check if Next button is clickable
    next_button = driver.find_element(By.ID, "paging-right")
    if not next_button.is_displayed():
        break

    # Click Next button
    next_button.click()

    # Add a delay to avoid overloading the server
    time.sleep(2)

# Clean up
driver.quit()
```

### Processed Data

```python
import os
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Specify the correct encoding (e.g., utf-8 or iso-8859-1)
file_encoding = "utf-8"
d = pd.DataFrame()
path = "submit/btphuongnhi/data/PIV/"
for filename in os.listdir(path):
    # print(filename)
    pathf = f'submit/btphuongnhi/data/PIV/{filename}'
    # Read the HTML content from the .xls file with the specified encoding
    with open(pathf, 'r', encoding=file_encoding) as file:
        html_content = file.read()

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find the table within the HTML content
    table = soup.find('table')

    # Convert the table to a pandas DataFrame
    df = pd.read_html(str(table))[0]
    d = pd.concat([d,df], ignore_index=True, axis=0)

# Print the DataFrame
d
```

```python
fdf = pd.DataFrame()
fdf["Ngày"] = d["Ngày"]["Ngày"]
fdf["Đóng cửa"] = d["Giá (nghìn VNĐ)"]["Đóng cửa"]
fdf["Điều chỉnh"] = d["Giá (nghìn VNĐ)"]["Điều chỉnh"]
fdf["Thay đổi"] = d["Thay đổi"]["Thay đổi"]
fdf["Khối lượng khớp lệnh"] = d["GD khớp lệnh"]["Khối lượng"]
fdf["Giá trị khớp lệnh (tỷ VNĐ)"] = d["GD khớp lệnh"]["Giá trị (tỷ VNĐ)"]
fdf["Khối lượng thỏa thuận"] = d["GD thỏa thuận"]["Khối lượng"]
fdf["Giá trị thỏa thuận (tỷ VNĐ)"] = d["GD thỏa thuận"]["Giá trị (tỷ VNĐ)"]
fdf["Mở cửa"] = d["Giá (nghìn VNĐ)"]["Mở cửa"]
fdf["Cao nhất"] = d["Giá (nghìn VNĐ)"]["Cao nhất"]
fdf["Thấp nhất"] = d["Giá (nghìn VNĐ)"]["Thấp nhất"]
```
